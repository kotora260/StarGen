```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ptick
```

```python
left = ['m400', 'm1000', 'm2200', 'm2800']
myr = [0.5702, 1.104, 1.703, 2.188]
```

```python
# plot the number of structures in 2D

plt.plot(left, [4, 4, 4, 5], marker='o', label='leaves')
plt.plot(left, [3, 3, 3, 4], marker='o', label='branches')

plt.title("The number of structures in 2D")
plt.xlabel("time progress")
plt.ylabel("number")
plt.legend()
```

```python
# make histgram the size of all structures in 2D

leaf_size_400 = [111.140978, 38.324475, 116.889650, 59.402937]
leaf_size_1000 = [49.821818, 40.240699, 42.156923, 136.051887]
leaf_size_2200 = [42.156923, 40.240699, 86.230069, 40.240699]
leaf_size_2800 = [548.039997, 49.821818, 53.654265, 76.648951, 45.989370]

branch_size_400 = [1458.246285, 444.563913, 249.109089]
branch_size_1000 = [837.38978, 365.998739, 1052.006847]
branch_size_2200 = [937.033421, 869.965589, 222.281957]
branch_size_2800 = [653.432304, 76.648951, 845.054680, 277.852446]

bins = 12
range = (0, 1500)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8), sharex=True, sharey=True)
axes[0, 0].hist(leaf_size_400, bins=bins, range=range, label='leaves')
axes[0, 0].hist(branch_size_400, bins=bins, range=range, alpha=0.5, label='branches')
axes[0, 1].hist(leaf_size_1000, bins=bins, range=range, label='leaves')
axes[0, 1].hist(branch_size_1000, bins=bins, range=range, alpha=0.5, label='branches')
axes[1, 0].hist(leaf_size_2200, bins=bins, range=range, label='leaves')
axes[1, 0].hist(branch_size_2200, bins=bins, range=range, alpha=0.5, label='branches')
axes[1, 1].hist(leaf_size_2800, bins=bins, range=range, label='leaves')
axes[1, 1].hist(branch_size_2800, bins=bins, range=range, alpha=0.5, label='branches')

plt.yticks(np.arange(0, 5, 1))

axes[0, 0].set_title('0.57 Myr')
axes[0, 0].set_ylabel('number')
axes[0, 0].legend(fontsize=12)
axes[0, 1].set_title('1.1 Myr')
axes[0, 1].legend(fontsize=12)
axes[1, 0].set_title('1.7 Myr')
axes[1, 0].set_xlabel('size [pc$^2$]')
axes[1, 0].set_ylabel('number')
axes[1, 0].legend(fontsize=12)
axes[1, 1].set_title('2.2 Myr')
axes[1, 1].set_xlabel('size [pc$^2$]')
axes[1, 1].legend(fontsize=12)
```

```python
# make histgram of the mass of all structures in 2D

leaf_mass_400 = [70.985543, 26.658508, 86.102272, 40.098052]
leaf_mass_1000 = [36.250389, 33.517472, 26.447831, 126.941156]
leaf_mass_2200 = [19.259471, 36.774012, 92.688766, 39.145864]
leaf_mass_2800 = [385.661354, 31.374755, 34.026615, 80.032790, 42.108332]

branch_mass_400 = [758.126444, 288.289484, 172.542741]
branch_mass_1000 = [544.230679, 287.651790, 637.470066]
branch_mass_2200 = [593.174759, 565.297372, 203.970062]
branch_mass_2800 = [441.683330, 58.876422, 528.877061, 228.792701]

bins = 12
range = (0, 800)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8), sharex=True, sharey=True)
axes[0, 0].hist(leaf_mass_400, bins=bins, range=range, label='leaves')
axes[0, 0].hist(branch_mass_400, bins=bins, range=range, alpha=0.5, label='branches')
axes[0, 1].hist(leaf_mass_1000, bins=bins, range=range, label='leaves')
axes[0, 1].hist(branch_mass_1000, bins=bins, range=range, alpha=0.5, label='branches')
axes[1, 0].hist(leaf_mass_2200, bins=bins, range=range, label='leaves')
axes[1, 0].hist(branch_mass_2200, bins=bins, range=range, alpha=0.5, label='branches')
axes[1, 1].hist(leaf_mass_2800, bins=bins, range=range, label='leaves')
axes[1, 1].hist(branch_mass_2800, bins=bins, range=range, alpha=0.5, label='branches')

plt.yticks(np.arange(0, 5, 1)) 

axes[0, 0].set_title('0.57 myr')
axes[0, 0].set_ylabel('number')
axes[0, 0].legend(fontsize=12)
axes[0, 1].set_title('1.1 Myr')
axes[0, 1].legend(fontsize=12)
axes[1, 0].set_title('1.7 Myr')
axes[1, 0].set_xlabel('mass [$M_$]')
axes[1, 0].set_ylabel('number')
axes[1, 0].legend(fontsize=12)
axes[1, 1].set_title('1.1 Myr')
axes[1, 1].set_xlabel('mass [$M_$]')
axes[1, 1].legend(fontsize=12)
```

```python
# plot thesize of trunk in 2D

trunk_size = [np.max(branch_size_400), np.max(branch_size_1000), np.max(branch_size_2200), np.max(branch_size_2800)]

plt.plot(myr, trunk_size, marker="o")

plt.title("size of trunk")
plt.xlabel("time [Myr]")
plt.ylabel('size [pc$^2$]')
```

```python
# plot the mass and virial mass of trunk in 2D

trunk_mass = [np.max(branch_mass_400), np.max(branch_mass_1000), np.max(branch_mass_2200), np.max(branch_mass_2800)]
fig, ax = plt.subplots()

ax.plot(myr, trunk_mass, marker="o", label='M')

plt.title("mass of trunk")
plt.xlabel("time [Myr]")
plt.ylabel('mass [$M_$]')
```
