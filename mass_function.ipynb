```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d400 = np.load("datasets/400.npz")
d1000 = np.load("datasets/1000.npz")
d2200 = np.load("datasets/2200.npz")
d2800 = np.load("datasets/2800.npz")
```

```python
m400 = d400["arr_2"]
m1000 = d1000["arr_2"]
m2200 = d2200["arr_2"]
m2800 = d2800["arr_2"]
```

```python
flat_m400 = m400.flatten()
flat_m1000 = m1000.flatten()
flat_m2200 = m2200.flatten()
flat_m2800 = m2800.flatten()
```

```python
plt.hist(flat_m400, bins=100, range=(0,350), log=True)

plt.title("m400")
plt.xlabel("mass")
plt.ylabel("number")
```

```python
plt.hist(flat_m1000, bins=100, range=(0,350), log=True)

plt.title("m1000")
plt.xlabel("mass")
plt.ylabel("number")
```

```python
plt.hist(flat_m2200, bins=100, range=(0,350), log=True)

plt.title("m2200")
plt.xlabel("mass")
plt.ylabel("number")
```

```python
plt.hist(flat_m2800, bins=100, range=(0,350), log=True)

plt.title("m2800")
plt.xlabel("mass")
plt.ylabel("number")
```

```python
bins = 70
range = (0, 350)

p400 = plt.hist(flat_m400, bins=bins, range=range, log=True, histtype='step', label="m400")
p1000 = plt.hist(flat_m1000, bins=bins, range=range, log=True, histtype='step', label="m1000")
p2200 = plt.hist(flat_m2200, bins=bins, range=range, log=True, histtype='step', label="m2200")
p2800 = plt.hist(flat_m2800, bins=bins, range=range, log=True, histtype='step', label="m2800")
plt.xscale("log")

plt.title("mass function")
plt.xlabel("mass [$M_☉$]")
plt.ylabel("number of voxels")
plt.legend()

plt.savefig("mass_function")
```
