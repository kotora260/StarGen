```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
step = ['m400', 'm1000', 'm2200', 'm2800']
myr = [0.5702, 1.104, 1.703, 2.188]

m400_leaves = pd.read_csv('datasets/mass_ppv/results_leaves_400_3D.csv')
m1000_leaves = pd.read_csv('datasets/mass_ppv/results_leaves_1000_3D.csv')
m2200_leaves = pd.read_csv('datasets/mass_ppv/results_leaves_2200_3D.csv')
m2800_leaves = pd.read_csv('datasets/mass_ppv/results_leaves_2800_3D.csv')
m400_branches = pd.read_csv('datasets/mass_ppv/results_branches_400_3D.csv')
m1000_branches = pd.read_csv('datasets/mass_ppv/results_branches_1000_3D.csv')
m2200_branches = pd.read_csv('datasets/mass_ppv/results_branches_2200_3D.csv')
m2800_branches = pd.read_csv('datasets/mass_ppv/results_branches_2800_3D.csv')
```

```python
leaves = [m400_leaves.shape[0], m1000_leaves.shape[0], m2200_leaves.shape[0], m2800_leaves.shape[0]]
branches = [m400_branches.shape[0], m1000_branches.shape[0], m2200_branches.shape[0], m2800_branches.shape[0]]

plt.plot(step, leaves, marker='o', label='leaves')
plt.plot(step, branches, marker='o', label='branches')
plt.yticks(np.arange(0, 6, 1)) 

plt.title("The number of structures in 2D")
plt.xlabel("time progress")
plt.ylabel("number")
plt.legend()
```

```python
fig = plt.figure(figsize=(20,20))
gs = fig.add_gridspec(4, 4)

bins = 5

ax1 = fig.add_subplot(gs[0, 0])
ax1.hist(m400_leaves['radius'], bins=bins, alpha=0.3, label='simulation 0.57 Myr')
ax1.hist(m1000_leaves['radius'], bins=bins, alpha=0.3, label='simulation 1.1 Myr')
ax1.hist(m2200_leaves['radius'], bins=bins, alpha=0.3, label='simulation 1.7 Myr')
ax1.hist(m2800_leaves['radius'], bins=bins, alpha=0.3, label='simulation 2.2 Myr')
ax1.hist(m400_branches['radius'], bins=bins, histtype='step', label='simulation 0.57 Myr')
ax1.hist(m1000_branches['radius'], bins=bins, histtype='step', label='simulation 1.1 Myr')
ax1.hist(m2200_branches['radius'], bins=bins, histtype='step', label='simulation 1.7 Myr')
ax1.hist(m2800_branches['radius'], bins=bins, histtype='step', label='simulation 2.2 Myr')
ax1.set_xlabel('radius [pc]')
ax1.set_ylabel('count')
ax1.legend(bbox_to_anchor=(0, -0.2), loc='upper left')

ax2 = fig.add_subplot(gs[0, 1])
ax2.scatter(m400_leaves['radius'], m400_leaves['mass'], marker='o')
ax2.scatter(m1000_leaves['radius'], m1000_leaves['mass'], marker='o')
ax2.scatter(m2200_leaves['radius'], m2200_leaves['mass'], marker='o')
ax2.scatter(m2800_leaves['radius'], m2800_leaves['mass'], marker='o')
ax2.scatter(m400_branches['radius'], m400_branches['mass'], marker='x')
ax2.scatter(m1000_branches['radius'], m1000_branches['mass'], marker='x')
ax2.scatter(m2200_branches['radius'], m2200_branches['mass'], marker='x')
ax2.scatter(m2800_branches['radius'], m2800_branches['mass'], marker='x')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel('radius [pc]')
ax2.set_ylabel('mass [$M_\odot$]')

ax3 = fig.add_subplot(gs[0, 2])
ax3.scatter(m400_leaves['radius'], m400_leaves['velocity dispersion'], marker='o')
ax3.scatter(m1000_leaves['radius'], m1000_leaves['velocity dispersion'], marker='o')
ax3.scatter(m2200_leaves['radius'], m2200_leaves['velocity dispersion'], marker='o')
ax3.scatter(m2800_leaves['radius'], m2800_leaves['velocity dispersion'], marker='o')
ax3.scatter(m400_branches['radius'], m400_branches['velocity dispersion'], marker='x')
ax3.scatter(m1000_branches['radius'], m1000_branches['velocity dispersion'], marker='x')
ax3.scatter(m2200_branches['radius'], m2200_branches['velocity dispersion'], marker='x')
ax3.scatter(m2800_branches['radius'], m2800_branches['velocity dispersion'], marker='x')
ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_xlabel('radius [pc]')
ax3.set_ylabel('velocity dispersion [km/s]')

ax4 = fig.add_subplot(gs[0, 3])
ax4.scatter(m400_leaves['radius'], m400_leaves['virial parameter'], marker='o')
ax4.scatter(m1000_leaves['radius'], m1000_leaves['virial parameter'], marker='o')
ax4.scatter(m2200_leaves['radius'], m2200_leaves['virial parameter'], marker='o')
ax4.scatter(m2800_leaves['radius'], m2800_leaves['virial parameter'], marker='o')
ax4.scatter(m400_branches['radius'], m400_branches['virial parameter'], marker='x')
ax4.scatter(m1000_branches['radius'], m1000_branches['virial parameter'], marker='x')
ax4.scatter(m2200_branches['radius'], m2200_branches['virial parameter'], marker='x')
ax4.scatter(m2800_branches['radius'], m2800_branches['virial parameter'], marker='x')
ax4.set_xscale('log')
ax4.set_yscale('log')
ax4.set_xlabel('radius [pc]')
ax4.set_ylabel('virial parameter')

ax6 = fig.add_subplot(gs[1, 1])
ax6.hist(m400_leaves['mass'], bins=bins, alpha=0.3)
ax6.hist(m1000_leaves['mass'], bins=bins, alpha=0.3)
ax6.hist(m2200_leaves['mass'], bins=bins, alpha=0.3)
ax6.hist(m2800_leaves['mass'], bins=bins, alpha=0.3)
ax6.hist(m400_branches['mass'], bins=bins, histtype='step')
ax6.hist(m1000_branches['mass'], bins=bins, histtype='step')
ax6.hist(m2200_branches['mass'], bins=bins, histtype='step')
ax6.hist(m2800_branches['mass'], bins=bins, histtype='step')
ax6.set_xlabel('mass [$M_\odot$]')
ax6.set_ylabel('count')

ax7 = fig.add_subplot(gs[1, 2])
ax7.scatter(m400_leaves['mass'], m400_leaves['velocity dispersion'], marker='o', label='simulation 0.57 Myr')
ax7.scatter(m1000_leaves['mass'], m1000_leaves['velocity dispersion'], marker='o', label='simulation 1.1 Myr')
ax7.scatter(m2200_leaves['mass'], m2200_leaves['velocity dispersion'], marker='o', label='simulation 1.7 Myr')
ax7.scatter(m2800_leaves['mass'], m2800_leaves['velocity dispersion'], marker='o', label='simulation 2.2 Myr')
ax7.scatter(m400_branches['mass'], m400_branches['velocity dispersion'], marker='x', label='simulation 0.57 Myr')
ax7.scatter(m1000_branches['mass'], m1000_branches['velocity dispersion'], marker='x', label='simulation 1.1 Myr')
ax7.scatter(m2200_branches['mass'], m2200_branches['velocity dispersion'], marker='x', label='simulation 1.7 Myr')
ax7.scatter(m2800_branches['mass'], m2800_branches['velocity dispersion'], marker='x', label='simulation 2.2 Myr')
ax7.set_xscale('log')
ax7.set_yscale('log')
ax7.set_xlabel('mass [$M_\odot$]')
ax7.set_ylabel('velocity dispersion [km/s]')
ax7.legend(bbox_to_anchor=(-2.4, 0.4), loc='upper left')

ax8 = fig.add_subplot(gs[1, 3])
ax8.scatter(m400_leaves['mass'], m400_leaves['virial parameter'], marker='o')
ax8.scatter(m1000_leaves['mass'], m1000_leaves['virial parameter'], marker='o')
ax8.scatter(m2200_leaves['mass'], m2200_leaves['virial parameter'], marker='o')
ax8.scatter(m2800_leaves['mass'], m2800_leaves['virial parameter'], marker='o')
ax8.scatter(m400_branches['mass'], m400_branches['virial parameter'], marker='x')
ax8.scatter(m1000_branches['mass'], m1000_branches['virial parameter'], marker='x')
ax8.scatter(m2200_branches['mass'], m2200_branches['virial parameter'], marker='x')
ax8.scatter(m2800_branches['mass'], m2800_branches['virial parameter'], marker='x')
ax8.set_xscale('log')
ax8.set_yscale('log')
ax8.set_xlabel('mass [$M_\odot$]')
ax8.set_ylabel('virial parameter')

ax11 = fig.add_subplot(gs[2, 2])
ax11.hist(m400_leaves['velocity dispersion'], bins=bins, alpha=0.3)
ax11.hist(m1000_leaves['velocity dispersion'], bins=bins, alpha=0.3)
ax11.hist(m2200_leaves['velocity dispersion'], bins=bins, alpha=0.3)
ax11.hist(m2800_leaves['velocity dispersion'], bins=bins, alpha=0.3)
ax11.hist(m400_branches['velocity dispersion'], bins=bins, histtype='step')
ax11.hist(m1000_branches['velocity dispersion'], bins=bins, histtype='step')
ax11.hist(m2200_branches['velocity dispersion'], bins=bins, histtype='step')
ax11.hist(m2800_branches['velocity dispersion'], bins=bins, histtype='step')
ax11.set_xlabel('velocity dispersion [km/s]')
ax11.set_ylabel('count')

ax12 = fig.add_subplot(gs[2, 3])
ax12.scatter(m400_leaves['velocity dispersion'], m400_leaves['virial parameter'], marker='o')
ax12.scatter(m1000_leaves['velocity dispersion'], m1000_leaves['virial parameter'], marker='o')
ax12.scatter(m2200_leaves['velocity dispersion'], m2200_leaves['virial parameter'], marker='o')
ax12.scatter(m2800_leaves['velocity dispersion'], m2800_leaves['virial parameter'], marker='o')
ax12.scatter(m400_branches['velocity dispersion'], m400_branches['virial parameter'], marker='x')
ax12.scatter(m1000_branches['velocity dispersion'], m1000_branches['virial parameter'], marker='x')
ax12.scatter(m2200_branches['velocity dispersion'], m2200_branches['virial parameter'], marker='x')
ax12.scatter(m2800_branches['velocity dispersion'], m2800_branches['virial parameter'], marker='x')
ax12.set_xscale('log')
ax12.set_yscale('log')
ax12.set_xlabel('velocity dispersion [km/s]')
ax12.set_ylabel('virial parameter')

ax16 = fig.add_subplot(gs[3, 3])
ax16.hist(m400_leaves['virial parameter'], bins=bins, alpha=0.3)
ax16.hist(m1000_leaves['virial parameter'], bins=bins, alpha=0.3)
ax16.hist(m2200_leaves['virial parameter'], bins=bins, alpha=0.3)
ax16.hist(m2800_leaves['virial parameter'], bins=bins, alpha=0.3)
ax16.hist(m400_branches['virial parameter'], bins=bins, histtype='step')
ax16.hist(m1000_branches['virial parameter'], bins=bins, histtype='step')
ax16.hist(m2200_branches['virial parameter'], bins=bins, histtype='step')
ax16.hist(m2800_branches['virial parameter'], bins=bins, histtype='step')
ax16.set_xlabel('virial parameter')
ax16.set_ylabel('count')

plt.show()
```

```python
mean_size_leaves = [m400_leaves['radius'].mean(), 
                    m1000_leaves['radius'].mean(), 
                    m2200_leaves['radius'].mean(), 
                    m2800_leaves['radius'].mean()]
mean_size_branches = [m400_branches['radius'].mean(), 
                      m1000_branches['radius'].mean(), 
                      m2200_branches['radius'].mean(), 
                      m2800_branches['radius'].mean()]
mean_mass_leaves = [m400_leaves['mass'].mean(),
                    m1000_leaves['mass'].mean(),
                    m2200_leaves['mass'].mean(),
                    m2800_leaves['mass'].mean()]
mean_mass_branches = [m400_branches['mass'].mean(),
                      m1000_branches['mass'].mean(),
                      m2200_branches['mass'].mean(),
                      m2800_branches['mass'].mean()]
mean_velocity_dispersion_leaves = [m400_leaves['velocity dispersion'].mean(),
                                    m1000_leaves['velocity dispersion'].mean(),
                                    m2200_leaves['velocity dispersion'].mean(),
                                    m2800_leaves['velocity dispersion'].mean()]
mean_velocity_dispersion_branches = [m400_branches['velocity dispersion'].mean(),
                                      m1000_branches['velocity dispersion'].mean(),
                                      m2200_branches['velocity dispersion'].mean(),
                                      m2800_branches['velocity dispersion'].mean()]
mean_virial_parameter_leaves = [m400_leaves['virial parameter'].mean(),
                                m1000_leaves['virial parameter'].mean(),
                                m2200_leaves['virial parameter'].mean(),
                                m2800_leaves['virial parameter'].mean()]
mean_virial_parameter_branches = [m400_branches['virial parameter'].mean(),
                                  m1000_branches['virial parameter'].mean(),
                                  m2200_branches['virial parameter'].mean(),
                                  m2800_branches['virial parameter'].mean()]

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, mean_size_leaves, marker='o', label='leaves')
ax.plot(myr, mean_size_branches, marker='o', label='branches')
#ax.set_title('size')
ax.set_xlabel('time [Myr]', fontsize=18)
ax.set_ylabel('size [pc]', fontsize=18)
ax.tick_params(labelsize=18)
ax.legend(fontsize=18)
plt.show()

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, mean_mass_leaves, marker='o', label='leaves')
ax.plot(myr, mean_mass_branches, marker='o', label='branches')
#ax.set_title('mass')
ax.set_xlabel('time [Myr]', fontsize=18)
ax.set_ylabel('mass [$M_\odot$]', fontsize=18)
ax.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
ax.tick_params(labelsize=18)
ax.yaxis.offsetText.set_fontsize(18)
plt.show()

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, mean_velocity_dispersion_leaves, marker='o', label='leaves')
ax.plot(myr, mean_velocity_dispersion_branches, marker='o', label='branches')
#ax.set_title('velocity dispersion')
ax.set_xlabel('time [Myr]', fontsize=18)
ax.set_ylabel('velocity dispersion [km/s]', fontsize=18)
ax.tick_params(labelsize=18)
plt.show()

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, mean_virial_parameter_leaves, marker='o', label='leaves')
ax.plot(myr, mean_virial_parameter_branches, marker='o', label='branches')
#ax.set_title('virial parameter')
ax.set_xlabel('time [Myr]', fontsize=18)
ax.set_ylabel('virial parameter', fontsize=18)
ax.tick_params(labelsize=18)
plt.show()
```
