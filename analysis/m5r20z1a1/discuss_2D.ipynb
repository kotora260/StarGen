```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
myr = [0.47, 0.81, 0.99, 1.1, 1.3, 1.4, 1.5, 1.7, 1.8, 1.9, 2.0]

for i in range(200, 2400, 200):
    filename = f'C:/users/kotor/StarGen/datasets/simdata/m5r20z1a1/results_leaves_{i}.csv'
    globals()[f'm{i}_leaves'] = pd.read_csv(filename)
    filename = f'C:/users/kotor/StarGen/datasets/simdata/m5r20z1a1/results_branches_{i}.csv'
    globals()[f'm{i}_branches'] = pd.read_csv(filename)
    if globals()[f'm{i}_branches'].empty:
       globals()[f'm{i}_structures'] = globals()[f'm{i}_leaves']
    else:
       globals()[f'm{i}_structures'] = pd.concat([globals()[f'm{i}_leaves'], globals()[f'm{i}_branches']], axis=0, ignore_index=True)
```

```python
leaves, branches = [], []
for i in range(200, 2400, 200):
    leaves.append(globals()[f'm{i}_leaves'].shape[0])
    branches.append(globals()[f'm{i}_branches'].shape[0])

plt.plot(myr, leaves, marker='o', label='leaves')
plt.plot(myr, branches, marker='o', label='branches')
plt.yticks(np.arange(0, 5, 1))
plt.title('Number of leaves and branches')
plt.xlabel('time [Myr]')
plt.ylabel('number')
plt.legend()
```

```python
fig = plt.figure(figsize=(20, 20))
gs = fig.add_gridspec(4, 4)

bins = 5

ax1 = fig.add_subplot(gs[0, 0])
for i in range(200, 2400, 200):
    ax1.hist(globals()[f'm{i}_leaves']['radius'], bins=bins, alpha=0.3, label=f'm{i}')
    ax1.hist(globals()[f'm{i}_branches']['radius'], bins=bins, histtype='step', label=f'm{i}')
ax1.set_xlabel('radius [pc]')
ax1.set_ylabel('count')

ax2 = fig.add_subplot(gs[0, 1])
for i in range(200, 2400, 200):
    ax2.scatter(globals()[f'm{i}_leaves']['radius'], globals()[f'm{i}_leaves']['mass'], marker='o')
    ax2.scatter(globals()[f'm{i}_branches']['radius'], globals()[f'm{i}_branches']['mass'], marker='x')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel('radius [pc]')
ax2.set_ylabel('mass [$M_\odot$]')

ax3 = fig.add_subplot(gs[0, 2])
for i in range(200, 2400, 200):
    ax3.scatter(globals()[f'm{i}_leaves']['radius'], globals()[f'm{i}_leaves']['velocity dispersion'], marker='o')
    ax3.scatter(globals()[f'm{i}_branches']['radius'], globals()[f'm{i}_branches']['velocity dispersion'], marker='x')
ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_xlabel('radius [pc]')
ax3.set_ylabel('velocity dispersion [km/s]')

ax4 = fig.add_subplot(gs[0, 3])
for i in range(200, 2400, 200):
    ax4.scatter(globals()[f'm{i}_leaves']['radius'], globals()[f'm{i}_leaves']['virial parameter'], marker='o')
    ax4.scatter(globals()[f'm{i}_branches']['radius'], globals()[f'm{i}_branches']['virial parameter'], marker='x')
ax4.set_xscale('log')
ax4.set_yscale('log')
ax4.set_xlabel('radius [pc]')
ax4.set_ylabel('virial parameter')

ax6 = fig.add_subplot(gs[1, 1])
for i in range(200, 2400, 200):
    ax6.hist(globals()[f'm{i}_leaves']['mass'], bins=bins, alpha=0.3, label=f'm{i}')
    ax6.hist(globals()[f'm{i}_branches']['mass'], bins=bins, histtype='step', label=f'm{i}')
ax6.set_xlabel('mass [$M_\odot$]')
ax6.set_ylabel('count')

ax7 = fig.add_subplot(gs[1, 2])
for i in range(200, 2400, 200):
    ax7.scatter(globals()[f'm{i}_leaves']['mass'], globals()[f'm{i}_leaves']['velocity dispersion'], marker='o', label=f'm{i}')
    ax7.scatter(globals()[f'm{i}_branches']['mass'], globals()[f'm{i}_branches']['velocity dispersion'], marker='x', label=f'm{i}')
ax7.set_xscale('log')
ax7.set_yscale('log')
ax7.set_xlabel('mass [$M_\odot$]')
ax7.set_ylabel('velocity dispersion [km/s]')

ax8 = fig.add_subplot(gs[1, 3])
for i in range(200, 2400, 200):
    ax8.scatter(globals()[f'm{i}_leaves']['mass'], globals()[f'm{i}_leaves']['virial parameter'], marker='o')
    ax8.scatter(globals()[f'm{i}_branches']['mass'], globals()[f'm{i}_branches']['virial parameter'], marker='x')
ax8.set_xscale('log')
ax8.set_yscale('log')
ax8.set_xlabel('mass [$M_\odot$]')
ax8.set_ylabel('virial parameter')

ax11 = fig.add_subplot(gs[2, 2])
for i in range(200, 2400, 200):
    ax11.hist(globals()[f'm{i}_leaves']['velocity dispersion'], bins=bins, alpha=0.3, label=f'm{i}')
    ax11.hist(globals()[f'm{i}_branches']['velocity dispersion'], bins=bins, histtype='step', label=f'm{i}')
ax11.set_xlabel('velocity dispersion [km/s]')
ax11.set_ylabel('count')

ax12 = fig.add_subplot(gs[2, 3])
for i in range(200, 2400, 200):
    ax12.scatter(globals()[f'm{i}_leaves']['velocity dispersion'], globals()[f'm{i}_leaves']['virial parameter'], marker='o')
    ax12.scatter(globals()[f'm{i}_branches']['velocity dispersion'], globals()[f'm{i}_branches']['virial parameter'], marker='x')
ax12.set_xscale('log')
ax12.set_yscale('log')
ax12.set_xlabel('velocity dispersion [km/s]')
ax12.set_ylabel('virial parameter')

ax16 = fig.add_subplot(gs[3, 3])
for i in range(200, 2400, 200):
    ax16.hist(globals()[f'm{i}_leaves']['virial parameter'], bins=bins, alpha=0.3, label=f'm{i}')
    ax16.hist(globals()[f'm{i}_branches']['virial parameter'], bins=bins, histtype='step', label=f'm{i}')
ax16.set_xlabel('virial parameter')
ax16.set_ylabel('count')
```

```python
trunk_size, trunk_mass, trunk_velocity_dispersion, trunk_virial_parameter = [], [], [], []
for i in range(200, 2400, 200):
    trunk_index = globals()[f'm{i}_structures']['radius'].idxmax()
    trunk_size.append(globals()[f'm{i}_structures']['radius'][trunk_index])
    trunk_mass.append(globals()[f'm{i}_structures']['mass'][trunk_index])
    trunk_velocity_dispersion.append(globals()[f'm{i}_structures']['velocity dispersion'][trunk_index])
    trunk_virial_parameter.append(globals()[f'm{i}_structures']['virial parameter'][trunk_index])

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, trunk_size, marker='o', label='size')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('size [pc]')

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, trunk_mass, marker='o', label='mass')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('mass [$M_\odot$]')

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, trunk_velocity_dispersion, marker='o', label='velocity dispersion')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('velocity dispersion [km/s]')

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, trunk_virial_parameter, marker='o', label='virial parameter')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('virial parameter')
```

```python
for i in range(200, 2400, 200):
    trunk_index = globals()[f'm{i}_structures']['radius'].idxmax()
    globals()[f'inner_structure_{i}'] = globals()[f'm{i}_structures'].drop(trunk_index)

inner_mean_size, inner_mean_mass, inner_mean_velocity_dispersion, inner_mean_virial_parameter = [], [], [], []
for i in range(200, 2400, 200):
    inner_mean_size.append(globals()[f'inner_structure_{i}']['radius'].mean())
    inner_mean_mass.append(globals()[f'inner_structure_{i}']['mass'].mean())
    inner_mean_velocity_dispersion.append(globals()[f'inner_structure_{i}']['velocity dispersion'].mean())
    inner_mean_virial_parameter.append(globals()[f'inner_structure_{i}']['virial parameter'].mean())

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, inner_mean_size, marker='o', label='size')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('size [pc]')

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, inner_mean_mass, marker='o', label='mass')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('mass [$M_\odot$]')

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, inner_mean_velocity_dispersion, marker='o', label='velocity dispersion')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('velocity dispersion [km/s]')

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(myr, inner_mean_virial_parameter, marker='o', label='virial parameter')
ax.set_xlabel('time [Myr]')
ax.set_ylabel('virial parameter')
```
